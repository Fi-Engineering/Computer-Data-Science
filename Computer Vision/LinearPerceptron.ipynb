### Define linear perceptron class

```python
class PrimalPerceptron(object):
	def __init__(self, x, y, w = None, b = None, sumW=None, sumB=None):
		
		num_sample, num_dims = x.shape
	
		np.random.seed(0)
		####################################
		# TODO: YOUR CODE HEREï¼š init weights

    # Init weights and bias
		

		w = np.ones(x.shape[1])
		

		b = np.zeros(1)
	
		sumW = np.zeros(x.shape[1])
		sumB = np.zeros(1)
	

		####################################
		#if not w: w = None
		#if not b: b = None
		self.x, self.y, self.w, self.b = x, y, w, b
		self.sumW = sumW
		self.sumB = sumB
		self.lr = 0.1

		
	def predict(self):
		####################################
		# TODO: YOUR CODE HERE, forward

		#yPred = np.matmul(self.x, self.w) + self.b
		yPred = np.where(np.matmul(self.x, self.w) + self.b <0, -1, 1)
	
	
		yTrue = yPred.reshape((-1, 1))
	
    
		preds = yTrue;
		y_h = labels

		####################################
		#preds = None
		#y_hat = None
		return preds, y_h

	def update(self):
		####################################
		# TODO: YOUR CODE HERE, backward
	
	
		for i in range(self.x.shape[0]):
		
			if(self.y[i]*(self.x[i]@self.w + self.b) <=0 ):
			
				self.w += self.y[i] * self.x[i] # ValueError: setting an array element with a sequence.
			
				self.b += self.y[i]
			
			self.sumW += self.sumW + self.w
		
			self.sumB += self.sumB + self.b

		####################################
		# update the weights and bias
		return
```
